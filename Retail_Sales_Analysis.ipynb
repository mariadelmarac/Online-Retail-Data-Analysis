```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/online_retail.csv'
data = pd.read_csv(file_path)

print(data.head())
```

```python
data_cleaned = data.drop_duplicates()

data_cleaned = data_cleaned.dropna(subset='CustomerID')

print(data_cleaned.head())
```

```python
data_cleaned.isnull().sum()
```

```python
data_cleaned.duplicated().sum()
```

```python
data_cleaned.head()
```

```python
data_cleaned['TotalAmount'] = data_cleaned['Quantity'] * data_cleaned['UnitPrice']
```

```python
data_cleaned.head()
```

```python
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])
```

```python
data_cleaned.info()
```

```python
data_cleaned['Year'] = data_cleaned['InvoiceDate'].dt.year
data_cleaned['Month'] = data_cleaned['InvoiceDate'].dt.month
data_cleaned['Day'] = data_cleaned['InvoiceDate'].dt.day

data_cleaned.head()
```

```python
sales_by_year = data_cleaned.groupby('Year')['TotalAmount'].sum()
sales_by_year
```

```python
data_cleaned['Quarter'] = data_cleaned['InvoiceDate'].dt.quarter
```

```python
sales_by_quarter = data_cleaned.groupby(['Year', 'Quarter'])['TotalAmount'].sum()
sales_by_quarter
```

```python
data_cleaned['Semester'] = data_cleaned['Month'].apply(lambda x: 1 if x <= 6 else 2)
```

```python
sales_by_semester = data_cleaned.groupby(['Year', 'Semester'])['TotalAmount'].sum()
sales_by_semester
```

```python
sales_by_month = data_cleaned.groupby(['Year', 'Month'])['TotalAmount'].sum()
sales_by_month
```

```python
data_cleaned.head()
```

```python
total_returned = data_cleaned[data_cleaned['Quantity'] < 0]
total_returned
```

```python
total_returned = data_cleaned[data_cleaned['Quantity'] < 0].shape[0]
total_returned
```

```python
total_sold = data_cleaned[data_cleaned['Quantity'] > 0].shape[0]
total_sold

items_not_returned = total_sold - total_returned
items_not_returned
```

```python
labels = ['Items sold', 'Items returned']
sizes = [items_not_returned, total_returned]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
def categorize_total_amout(total_amount):
    if total_amount < 100:
        return 'Low'
    elif total_amount < 500:
        return 'Medium'
    else:
        return 'High'

data_cleaned['AmountCategory'] = data_cleaned['TotalAmount'].apply(categorize_total_amout)

data_cleaned.head()
```

```python
total_low = data_cleaned[data_cleaned['AmountCategory'] == 'Low'].shape[0]
total_medium = data_cleaned[data_cleaned['AmountCategory'] == 'Medium'].shape[0]
total_high = data_cleaned[data_cleaned['AmountCategory'] == 'High'].shape[0]

labels = ['Low', 'Medium', 'High']
sizes = [total_low, total_medium, total_high]
colors = ['lightblue', 'lightcoral', 'lightgreen']
explode = (0.1, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data_cleaned.groupby(['Year', 'Month'])['TotalAmount'].sum().plot(kind='bar')
plt.title('Total Sales by Month')
plt.xlabel('Year, Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
top_items = data_cleaned.groupby(['StockCode', 'Description'])['Quantity'].sum().sort_values(ascending=False).head(10)
top_items = top_items.reset_index()
top_items
```

```python
plt.figure(figsize=(12, 8))
plt.barh(top_items['Description'].astype(str).fillna('Unknown Description'), top_items['Quantity'])
plt.xlabel('Quantity')
plt.ylabel('Description')
plt.gca().invert_yaxis() # Add parentheses to call the function
plt.title('Top 10 Items by Quantity')
plt.show()
```

# Online Retail Data Analysis

This notebook performs an exploratory data analysis on an online retail dataset.

## Dataset

The dataset contains transactional data from an online retail store.

## Analysis Performed

1.  **Data Loading and Initial Inspection**: Loaded the data and checked for basic information, descriptive statistics, null values, and duplicates.
2.  **Data Cleaning**: Removed duplicate rows and rows with missing `CustomerID`.
3.  **Feature Engineering**:
    *   Calculated `TotalAmount` as the product of `Quantity` and `UnitPrice`.
    *   Converted `InvoiceDate` to datetime objects.
    *   Extracted `Year`, `Month`, `Day`, `Quarter`, and `Semester` from `InvoiceDate`.
4.  **Sales Analysis**:
    *   Calculated total sales by year, quarter, semester, and month.
5.  **Return Analysis**: Identified and counted returned items (where `Quantity` is negative).
6.  **Sales Category Analysis**: Categorized sales into 'Low', 'Medium', and 'High' based on `TotalAmount`.
7.  **Visualization**:
    *   Generated a pie chart showing the proportion of sold items versus returned items.
    *   Generated a pie chart showing the distribution of sales by amount category (Low, Medium, High).
    *   Plotted a bar chart showing the total sales by month.
    *   Identified and visualized the top 10 items by quantity sold using a horizontal bar chart.

## Key Findings

*   The dataset contains transactional data for a retail store.
*   There were a significant number of missing values in the `CustomerID` column and some in the `Description` column, which were handled by dropping rows with missing `CustomerID` and duplicate entries.
*   Negative quantities indicate returned items.
*   Sales show a clear seasonality, with a peak in November.
*   The majority of transactions fall into the 'Low' amount category.
*   The top-selling items by quantity were identified.

## Code Structure

The analysis is structured into several code cells, each performing a specific task:

*   Loading data and initial checks.
*   Cleaning the data.
*   Creating new features.
*   Analyzing sales by different time periods.
*   Analyzing returned items.
*   Categorizing sales amount.
*   Visualizing the findings.
